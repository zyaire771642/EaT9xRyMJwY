## **What is cloud computing**

The cloud services typically possess the following capabilities:





**1. Core capabilities: Rapid elasticity**

A typical e-commerce site will have a network traffic of 0.3 million requests on a normal day which can be handled by about 250 servers. However, during peak sale periods like promotional events, the traffic increases to around 1.5 million requests per day which needs around an additional 5000 servers.



**What is rapid elasticity**

Cloud services provide the consumer, with the option of quickly "stretching" and "releasing" the capacity of the computing resources as per their needs at any point of time. 

 

**2. Core capabilities: On-demand self-service**


When a small business, like a web design studio, wants to back up all their work related data for a short duration of time during an OS upgrade activity, they would normally contact their hardware vendor and place a request for required amount of temporary storage (hard disks).

**Traditional IT resource provisioning processes**


For augmentation and/or release of resources, like storage capacity, you would be required to
create service request tickets by filling up forms or calling the IT support center.
constantly follow up with the IT support team on the status of the tickets.


Also, mostly there is no direct communication between you the requestor and the technician who is servicing the requests. Your request goes through intermediate call centers, first level support team, multiple updates, etc. All these increase the chances of errors in service due to misinterpretations and miscommunications.

This would hinder the rapid elasticity of resources.

**Resource provisioning in a cloud service**

Here is a snapshot of the web based self-service portal provided by Google for its cloud-based storage solution - Google Drive, where the customers can keep switching between 15 GB, 100 GB or 1 TB storage based on their needs.



**3. Core capabilities: Resource pooling**

A research scientist might require a lot of processing power for data analysis and calculations whereas a blogger might not be able to purchase an expensive standalone webserver to design and host their blog site. Many times individuals' PCs might not have the required hardware or software configuration suitable for the task in hand.

XenDesktop is a cloud-based desktop virtualization service from Citrix, that allows multiple users to access and run virtual desktops, customized to their needs, that are created out of resources present in huge server farms at a centralized location (resource pool).

Though multiple virtual desktop instances might be hosted using the same hardware resources, each instance runs in isolation, as if it is an independent desktop.

Any configuration of virtual desktop can be provisioned for/modified, by carving out resources from the pool, as per the customer's requests.



**4. Core capabilities: Metered billing service**

A developer, who needs a month to build a simple application in .Net platform, would definitely need Microsoft Visual Studio suite.




**What is metered billing**

A cloud service provider measures and meters the usage of resources by a consumer. This is accomplished by including built-in services that tracks the quantity of various IT resources a consumer uses.

Usually, you need not pay upfront or commit to any minimum resource allocation. You must pay at the end of the usage period, only for the number of units of resources that were used. This model of cloud service billing is the most popular and its called Pay As You Go (PAYG).

### **What is IaaS?**

Infrastructure as a Service is the provisioning of IT infrastructure resources, like processors, storage, networks, firewalls, load balancers etc., over the internet.

**Popular IaaS offerings**



**How IaaS works**



->Depending on the business requirements as consumers, IaaS providers allocate hardware components (real or virtual) from their resource pool and make it available for use, over the internet.

->Charged based on the resource allocation/usage.

->The IaaS provider is the owner of the infrastructure and is responsible for housing, running and maintaining it.

->IaaS can be used to create profitable and easily scalable IT solutions by enterprise customers as the difficulties of managing the underlying hardware are outsourced to the cloud provider.

->If a customer's scale of business operations vary, or they are looking for an expansion, they can tap into the CSPs resource pool as and when they need it rather than purchasing, installing and integrating the hardware themselves.

## **What is PaaS?**

Platform as a Service provides all the capabilities that is required to support a complete application lifecycle such as building, testing, deploying, managing, and updating using the same integrated environment.

Popular PaaS offerings



**How PaaS works?**



->The platform, provided by the CSP, typically includes OS, development tools, programming-language execution environment (runtime), database management systems etc. aimed to help developers (consumers) build, test and host applications.

->PaaS environments, hosted on the cloud, can be accessed by developers (consumers) through their web browser.

->Customers can pick and choose features like software tools, IDEs, frameworks to customize their platform or opt for a pre-configured platform offered by the CSP.

->PaaS offerings not only consist of development platforms, but also include the required infrastructure services (IaaS) to support the platform.

->PaaS helps businesses to develop and deploy applications faster, resulting in cost savings. Most importantly, it allows them to focus on the innovation of their applications and business rather than dealing with the hassles of buying and managing software licenses, setting up configurations and managing servers and databases.

## **What is SaaS?**

Software as a Service describes any cloud service where a fully functional and complete software product is delivered to users through the internet. Rather than installing and maintaining the software, you can simply access it through Internet, freeing from complex hardware and software management.

**Popular SaaS offerings**



**How SaaS works?**



->The SaaS provider hosts the application and data centrally on their resources. Hence, the SaaS provider is responsible not only for  providing and managing the software but also the infrastructure and platform on which it runs.

->Most SaaS applications can be accessed through any internet enabled device. With these services, you log into your account over the internet, often from a web browser.

->In the case of traditional software applications you must purchase the software as a package and then install it in your computer. Also, the licence of the software may limit the number of users and/or devices where it can be deployed.

->However, in the case of Software as a Service, users subscribe to the software instead of purchasing it, usually on monthly basis. Applications are usually purchased and used online, data files generally stored in the cloud and not on individual computers.

->SaaS offerings are generally adopted by businesses that want to benefit from application usage without investing huge amounts on licenses.

**Distinguishing SaaS from hosted applications**

**Example of a hosted application**

If you want to create a blog, you could visit wordpress.org and download the software that you would need to start-off and maintain your own blog. You should upload the application to the web server and then run the installation processes. In this case, you

->are responsible for hosting and maintaining the web site

->can modify the website's source code whenever you like

->have the complete and exclusive control of all the data in your database. Nobody can delete your account or block your access to it.

->WordPress is a software product owned by you is hosted on your web server. Even though ,this approach means more work, you have more control over the application and the hosting environment.

**Example of a SaaS**

If you dont have access to a web server, you could go to wordpress.com and subscribe to one of their plans. WordPress will host your application for you. However,

->Wordpress organization uses their resources to store and host your data

->they retain control of all the data

->you cant change the source code or transfer it to another server

->The software effectively belongs to WordPress, and you are only making use of the services that this software offers.

->Even though you lose some control over the application and your data, you benefit a lot in terms of ease-of-use, convenience and Total Cost of Ownership(TCO).

**Consumer concerns**

Consumers looking at leveraging the advantages of cloud computing by adopting cloud services (IaaS, PaaS, SaaS), have a few concerns that needs to be addressed by the CSPs.

 


To address the concerns of cloud consumers, cloud service providers have come up with multiple ways (deployment models) using which they can make their services (IaaS, PaaS or SaaS) available to the cloud consumers. 

Each deployment model aims at addressing one or more concerns of the cloud consumer. Therefore, it is important that consumers prioritize their concerns before opting for a particular model.





## Public cloud


The public cloud is the one that is available for the general public use. Hence, it is the most common and popular deployment model available today.

Examples-

->DropBox provides storage space to the general public.

->Google provides Gmail and other cloud servers to the general public

**Characteristics of public cloud :**

->The public cloud infrastructure exists on the premises of the cloud service provider.

->They are entirely owned, deployed, monitored, and managed by the cloud service provider, who deliver their computing resources over the internet.

->Public clouds provide the same services to all its users using the same shared infrastructure.

->Clients of the public cloud can range from individual users to enterprises.

**Concerns addressed -**

Lack of resources/expertise 

Managing cloud spend

## Private cloud

A private cloud is available only to users within a single organization.

Example -

PrimeHealth, a private healthcare service provider, might want to migrate their data to cloud, to leverage the benefits of cloud computing. But considering the confidentiality of patient records and strict regulations, PrimeHealth does not want to use a third-party CSP. Instead, they choose to host their own servers and make these servers available only to their employees through the internet.

**Characteristics of private cloud :**

->In private cloud, IT services are provided over private IT infrastructure for the exclusive use of business units, belonging to a single organization. The private cloud resources sit safely within the corporate firewall, on or off premises. Hence private clouds are also called corporate, enterprise or internal clouds.

->Here, the same organization is both the cloud consumer and cloud provider, where
a separate department usually has the responsibility for providing the cloud (and therefore assumes the cloud provider role)
the business units requiring access to the private cloud assume the cloud consumer role

->The typical cost savings that come with public cloud computing are not realized as the organization owns and maintains the IT infrastructure themselves.

->The chief difference between a private and a public cloud is that the former incorporates an additional layer of security by leveraging an enterprise firewall. This makes a private cloud the most secure type of cloud computing environment.

**Concerns addressed :**

Security

Compliance

Governance/Control

Performance

## Community cloud

A community cloud is a private cloud that is shared by two or more organizations having shared concerns like security requirements, policy, and compliance considerations etc.

Example -

PrimeHealth, after setting up their private cloud, decides to share it with other healthcare providers in their community. IT infrastructure administrators at PrimeHealth can configure it as a community cloud so that employees at the other healthcare centers can also access it. This is now a community cloud. 

This is mutually beneficial, as the cost of owning and maintaining the private infrastructure is now shared with the other healthcare providers, who share the same concerns.

**Characteristics of community cloud :**

->A community cloud may be owned, managed, and operated by all the participating organizations or by a third-party provider, on or off premises.

->A community cloud model can produce greater cost savings than the private cloud whilst offering some of its security features.

**Concerns addressed :**

Security

Managing cloud spend

Compliance

Governance/Control

Performance

## Hybrid cloud

A hybrid cloud is a combination of two or more different cloud deployment models.

Example -

If PrimeHealth chooses to conduct a large-scale study involving complex data, they would have to rapidly scale up for the big project and then reduce capacity after it finishes. Instead of investing in hardware to accommodate the short-term need, they can send the extra workload to a public cloud and use its computing power to facilitate data analysis. The private cloud would continue supporting PrimeHealths regular operations.

**Characteristics of hybrid cloud :**

->In hybrid cloud, the different clouds are bound together at a network level but remain individual entities.

->This deployment model helps businesses to make use of secured applications and data hosting on a private cloud, while still enjoying rapid elasticity by keeping shared data and applications on the public cloud.

->When the existing private cloud infrastructure is facing demand spikes, the additional resources that are required by a particular application can be accessed from the public cloud. This is termed as cloud bursting and is possible with the hybrid cloud.

->Hybrid cloud helps organizations in maximizing their efficiency by making use of public cloud services for all the non-sensitive operations, relying on a private cloud only where they require it and make sure that all of their platforms are smoothly integrated.

**Concerns addressed :**

Security

Compliance

Governance/Control

Performance
